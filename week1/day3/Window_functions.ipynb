### 1.Assign a unique row number to all employees based on salary (highest first). 

```sql
SELECT emp_id,
       emp_name,
       salary,
       ROW_NUMBER() OVER (ORDER BY salary DESC) AS row_num
FROM empl;
```

### 2.Assign row numbers to employees within each department based on salary descending. 

```sql
SELECT emp_id,
       emp_name,
       department,
       salary,
       ROW_NUMBER() OVER (partition by department ORDER BY salary DESC) AS row_num
FROM empl;
```

### 3.Assign row numbers based on employee joining date (latest first). 

```sql
SELECT emp_id,
       emp_name,
       join_date,
       ROW_NUMBER() OVER (ORDER BY join_date DESC) AS row_num
FROM empl;
```

### 4.Assign row numbers within each department based on earliest joining date. 

```sql
SELECT *,
       ROW_NUMBER() OVER (partition by department ORDER BY join_date asc) AS row_num
FROM empl;
```

# RANK() ONLY Questions

### 1.Rank all employees based on salary (highest first). 

```sql
SELECT emp_name,
       salary,
       RANK() OVER (ORDER BY salary DESC) AS rank_num
FROM empl;
```

### 2.Rank employees within each department based on salary. 

```sql
SELECT *,
       RANK() OVER (partition by department ORDER BY salary DESC) AS rank_num
FROM empl;
```

### 3.Rank employees based on joining date (latest gets rank 1). 

```sql
SELECT emp_id,
       emp_name,
       join_date,
       RANK() OVER (ORDER BY join_date DESC) AS rank_num
FROM empl;
```

### 4.Rank employees within department based on salary (lowest first). 

```sql
SELECT *,
       RANK() OVER (partition by department ORDER BY salary ASC) AS rank_num
FROM empl;
```

## 🔹 DENSE_RANK() ONLY Questions

### Assign dense rank to employees based on salary (highest first). 

```sql
select *,
dense_rank() over(order by salary desc) as rank_num
from empl
```

### Assign dense rank within each department based on salary. 

```sql
SELECT *,
       dense_rank() OVER (partition by department ORDER BY salary DESC) AS rank_num
FROM empl;
```

### Assign dense rank to employees based on joining date. 

```sql
select *,
dense_rank() over(order by join_date) as rank_num
from empl
```
