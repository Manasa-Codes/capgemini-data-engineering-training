### From the column mixed_value, extract the numeric characters that appear at the very beginning of the value. The extraction must stop as soon as a non-numeric character appears.

```sql
select mixed_value,
REGEXP_SUBSTR(mixed_value, '^[0-9]+')
from regex_practice
```

### From the column mixed_value, extract the numeric characters that appear at the very end of the value.
### No alphabetic or special characters should be included.

```sql
select mixed_value ,
regexp_substr(mixed_value,'[0-9]+$')
from regex_practice
```

### From the column mixed_value, extract only the first single character of the value, regardless of whether it is a letter or a number.

```sql
select mixed_value,
regexp_substr(mixed_value,'^.')
from regex_practice
```

### From the column mixed_value, extract only the last single character of the value.

```sql
select mixed_value,
regexp_substr(mixed_value,'.$')
from regex_practice
```

### From the column mixed_value, extract exactly two consecutive numeric characters that appear
### anywhere in the value.
### Do not extract more or fewer than two digits.

```sql
select mixed_value,
regexp_substr(mixed_value,'[0-9]{2}')
from regex_practice
```

### From the column mixed_value, extract exactly one numeric character that appears anywhere in the value.

```sql
select mixed_value,
regexp_substr(mixed_value,'[0-9]')
from regex_practice
```

### From the column mixed_value, extract the numeric portion that is present between alphabetic characters.

```sql
SELECT 
  mixed_value,
  REGEXP_SUBSTR(mixed_value, '(?<=[A-Za-z])[0-9]+(?=[A-Za-z])') AS extracted_number
FROM regex_practice
```

### From the column email, extract the text that appears before the at-symbol. The extracted value must not include the at-symbol itself.

```sql

select REGEXP_SUBSTR(email,'[^@]+') as extracted_mail
from regex_practice
```

### From the column email, extract the text that appears after the at-symbol including the domain name.

```sql
select REGEXP_SUBSTR(email,'@(.+)') as extracted_mail
from regex_practice
     
```

### From the column email, extract only the domain name without including the at-symbol.

```sql

SELECT 
  email,
  REGEXP_SUBSTR(email, '(?<=@).+') AS domain
FROM regex_practice
```

### From the column email, extract only the text that appears after the last dot in the email address.

```sql

select REGEXP_SUBSTR(email,'[^.]+$') as extracted_mail
from regex_practice
```

### From the column mixed_value, extract only alphabetic characters that appear together as a continuous sequence.

```sql
select REGEXP_SUBSTR(mixed_value,'[A-Za-z]+') as chars
from regex_practice
```

### Q14 From the column mixed_value, extract only numeric characters that appear together as a continuous sequence.

```sql

select REGEXP_SUBSTR(mixed_value,'[0-9]+') as chars
from regex_practice
```

### From the column full_text, extract exactly the first three characters of the value.

```sql

select REGEXP_SUBSTR(full_text,'^.{3}') as chars
from regex_practice
```

### From the column full_text, extract exactly the last two characters of the value.

```sql

select REGEXP_SUBSTR(full_text,'.{2}$') as chars
from regex_practice
```

### From the column full_text, extract the employee number portion that appears between the alphabetic prefix and the first underscore.

```sql

SELECT 
  full_text,
  REGEXP_EXTRACT(full_text, '^[A-Za-z]+([0-9]+)', 1) AS emp_number
FROM regex_practice
     
```

### From the column full_text, extract the country code that appears at the end of the value.

```sql
select REGEXP_SUBSTR(full_text,'[0-9]+$') as country_code
from regex_practice
```

### From the column full_text, extract the alphabetic text that appears between two underscore characters.

```sql

select REGEXP_SUBSTR(full_text,'_([A-Za-z]+)_') as alph_text
from regex_practice
```
