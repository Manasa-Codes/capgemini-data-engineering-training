```sql
CREATE TABLE Em (
    emp_id INT PRIMARY KEY,
    name VARCHAR(50),
    department_id INT,
    salary INT,
    manager_id INT
);
```

```sql
CREATE TABLE Dep (
    dept_id INT PRIMARY KEY,
    dept_name VARCHAR(50)
);
```

```sql
CREATE TABLE Ord (
    order_id INT PRIMARY KEY,
    emp_id INT,
    order_amount INT,
    order_date DATE
);
```

# Subquery Questions (Basic – 5)

### 1.Find employees who earn more than average salary

```sql
SELECT * FROM Em
WHERE salary>(SELECT AVG(salary) FROM Em)
```

### 2.Find employees who belong to the same department as 'Alice'

```sql
  SELECT * FROM Em
  WHERE department_id=(
    SELECT department_id FROM Em WHERE name='Alice'  );
```

### 3.Get employees whose salary is equal to the minimum salary

```sql
SELECT * FROM Em
WHERE salary=(SELECT MIN(salary) FROM Em)
```

### 4.Find employees who have placed at least one order


```sql
SELECT *
FROM Em e
WHERE EXISTS (
    SELECT 1
    FROM Ord o
    WHERE o.emp_id = e.emp_id
);
```

### 5.Get employees whose salary is greater than ALL employees in department 102


```sql
SELECT * FROM Em
where salary=(select max(salary) from Em where department_id=102)
```

#  CTE (Common Table Expression) Questions (Basic – 5)

### 1.Create a CTE to show employees with salary > 50,000 and fetch all records


```sql
WITH HighSalaryEmployees AS (
    SELECT *
    FROM Em
    WHERE salary > 50000
)
SELECT *
FROM HighSalaryEmployees;
```

### 2.Find average salary per department using CTE


```sql
WITH AvgSalary AS(
  SELECT department_id,AVG(salary) AS avg_salary
  FROM Em
  GROUP BY department_id
)
SELECT *
FROM AvgSalary;
```

### 3.Use CTE to get employees and their department names


```sql
WITH EmpDep AS (
  SELECT e.emp_id, e.name, e.department_id, e.salary, d.dept_name AS department_name
  FROM Em e
  JOIN Dep d ON e.department_id = d.dept_id
)
SELECT name, department_name
FROM EmpDep
```

### 4.Find total order amount per employee using CTE

```sql
WITH TotalAmount AS (
  SELECT e.emp_id, e.name AS employee, SUM(o.order_amount) AS total_amount
  FROM Em e
  JOIN Ord o ON o.emp_id = e.emp_id
  GROUP BY e.emp_id,e.name
)
SELECT *
FROM TotalAmount
```

### 5.Find employees whose salary is above department average using CTE

```sql
WITH DeptAvgSalary AS (
    SELECT department_id, AVG(salary) AS avg_salary
    FROM Em
    GROUP BY department_id
)
SELECT e.*
FROM Em e
JOIN DeptAvgSalary d
ON e.department_id = d.department_id
WHERE e.salary > d.avg_salary;
```
